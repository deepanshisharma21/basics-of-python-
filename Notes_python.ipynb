## **Data types** : A data type classifies the kind of value a variable can hold. It tells the Python interpreter how to store and manage the data. This means you can perform mathematical operations on numbers but not on text, which prevents unexpected errors. For example, Python recognizes 42 as a number and "Python" as text.

**Python's Built-in Data Types**
Python provides several standard data types ready for you to use. These types fall into several categories:
There are two types of data-types in python in terms of layman language.

1.Primary Data type
2.Secondary Data type or Container Data type

**Primary Data type :**
It consists of the following data types :
*  Int or Integer
* Float or decimal numbers
* String
* None data type
* Boolean data type

**Secondary Data type :** It contains primary data types (the ones mentioned above) as it's elements or values, basically they are container data types that consist of various primary data type values stored in them which allows us to store multiple values of any data type and assign to just one variable. This reduces of stating multiple variables for similar values.

* Lists (Arrays variation of C and C++ Lang for python)
* Tuples (Basically immutable python lists)
* Sets
* Dictionaries (Comsists of keys mapped to values, unordered (don't follow any order such as index, keys&Values can be in any order)
Collections (ordered dictionaries)
* Numpy, etc (Numerical arrays for scientific and machine learning projects)
All these data types have unlimited memory allocation (which implies it depends on one's PC memory limits, this also means you don't need to assign size of array, string, etc like in case of C/C++ as python has unlimited memory allocation and also automated garbage collector which works hand in hand with dynamic typing.



> To identify the type of any variable or object in python, use the type() function







```python
# --- Primary Data Types ---

# Integer (int)
my_integer = 10
print("Integer:", my_integer ,type(my_integer))

# Float (float)
my_float = 3.14
print("Float:",my_float,type(my_float))

# String (str)
my_string= "Hello,Python!" #strings are represented by double quotes "" or '' always
print("String:",my_string,type(my_string))

# Boolean (bool)
my_boolean_true = True #REMEMBER T IS CAPITAL AS WELL AS F IN TRUE,FALSE
my_boolean_false = False
print(f"Boolean (True): {my_boolean_true}, Type: {type(my_boolean_true)}")
print(f"Boolean (False): {my_boolean_false}, Type: {type(my_boolean_false)}")

# NoneType (None)
my_none = None
print(my_none,type(my_none))

# --- Secondary (Container) Data Types ---

# List (list) - ordered, changeable, allows duplicate members
my_list = [1, 2, 3, "apple", 3.14]
print(f"List: {my_list}, Type: {type(my_list)}")

# Tuple (tuple) - ordered, unchangeable, allows duplicate members
my_tuple = (1, 2, 3, "banana", 3.14)
print(f"Tuple: {my_tuple}, Type: {type(my_tuple)}")

# Set (set) - unordered, unindexed, no duplicate members
my_set = {1, 2, 3, "orange", 1}
print(f"Set: {my_set}, Type: {type(my_set)}")

# Dictionary (dict) - unordered, changeable, indexed (by keys), no duplicate keys
my_dictionary = {"name": "Alice", "age": 30, "city": "New York"}
print("Dictionary:",my_dictionary, "Type:",type(my_dictionary))
```

Indexing and slicing strings in python :
***Definition of Index/Indices*** :

It is a common method for keeping track of data so that it can be accessed quickly. Like an index in a book, it is a numeric value in which each entry contains the name of the item or element and its location. However, computer-based indexes may point to a physical location on a disk or to a logical location that points elsewhere to the actual location.
Indices (plural of 'index') start from 0 (zero) and go on to n-1, where 'n' is the number of characters in the string in this case. For eg: the string "hello" has n = 5 value as "hello" has 5 characters, hence the indices will start from 0 and go on till n-1, i.e, 5 - 1 = 4.

---
In python, indexing is represented as follows, say : a = "hello"; then a[index_number] is the syntax of using indices for string (as well as lists which will be explained later). So in this case, a[0], a[1], a[2], a[3] and a[4] are indices of the variable 'a'. Each of these indices, hold a character of that string assigned to the variable. Therefore, for a = "hello", a[0]=>'h', a[1]=>'e',a[2]=>'l', a[3]=>'l', a[4]=>'o'. We don't have a[5] as the length of "hello" is 5, and indices range from 0 to n-1 as I said before, so here it is 0 to 4 indices. Similar example : b = "Name", here indices of 'b' from 0 to 3 where, b[0]='N', b[1]='a', b[2]='m' and b[3]='e'.



```python
a = "apple"
print (" ",a[0])                #Positive indices from 0 to 4 (coz "apple" has 5 characters so, it's from 0 to n-1, or 5-1 = 4)
print (" ",a[1],a[4])
print (" ",a[2],"",a[3])
print (" ",a[3]," ",a[2])
print (" ",a[4],"  ",a[1])
print ("---------",a)
print (" ",a[-1],"  ",a[-4])    #Negative indices from -1 to -5 (coz "apple" has 5 characters)
print (" ",a[-2]," ",a[-3])
print (" ",a[-3],"",a[-2])
print (" ",a[-4],a[-1])
print (" ",a[-5])
```

#Recursion:
When a function calls itself repeatedly
It is a type of loops only,
all loops can have recursion ,all recursion can have loops.

```python
def show(n):
  if(n==0):
     return
  print(n)
  show(n-1)
show(5)
```

```python
def fact(n):
  if (n == 1 or n == 0):
      return 1
  return fact(n-1) * n

print(fact(5))
#RECURSION EXAMPLE

```

#DECORATORS
A decorator is a function that takes another function and adds extra functionality without changing the original function.
Syntax: @decorator_name

```python
def extra(func):
    def wrapper():
        print("Before")
        func()
        print("After")
    return wrapper

@extra
def say_hello():
    print("Hello")

say_hello()

#If decorator present, the return of the decorator will be executed
```

collections Module:
The collections module provides specialized data types that improve upon standard lists, tuples, and dictionaries.


```python
#1. Counter

#Counts the frequency of elements.
from collections import Counter

data = [1, 2, 2, 3, 3, 3]
c = Counter(data)
print(c)
```

```python
#2. defaultdict
#Provides a default value if a key does not exist.
from collections import defaultdict

d = defaultdict(int)
d['a'] += 1
print(d)
```

```python
#3. deque
#Allows fast insertion and deletion from both ends.
from collections import deque

dq = deque([1, 2, 3])
dq.append(4)
dq.appendleft(0)
print(dq)
```

```python
#4. namedtuple
#Creates tuple-like objects with named fields.
from collections import namedtuple

Student = namedtuple('Student', ['name', 'age'])
s = Student('Deepanshi', 18)
print(s.name, s.age)
```

itertools Module:
The itertools module provides fast, memory-efficient tools for iteration.

```python
#count()
#Generates infinite numbers.
from itertools import count

for i in count(1):
    if i == 5:
        break
    print(i)
```

```python
#2. cycle()

#Repeats elements infinitely.
from itertools import cycle

c = cycle(['A', 'B'])
for i in range(4):
    print(next(c))
```

```python
#3. permutations()

#Returns all possible ordered arrangements.
from itertools import permutations

print(list(permutations([1, 2, 3], 2)))
```

```python
#4.combinations()
#Returns all possible unordered combinations.
from itertools import combinations

print(list(combinations([1, 2, 3], 2)))

```

Lambda Functions:

Lambda functions are small anonymous functions written in one line.

```python
add = lambda a, b: a + b
print(add(2, 3))
```

```python
nums = {"a":1, "b":2, "c":3, "d":4}

print(nums.keys())
```

```python
#Lambda with map()
#Applies a function to each element.
nums = {"a":1, "b":2, "c":3, "d":4}

# Get the squared values using map and lambda
squared_values = map(lambda x: x**2, nums.values())

# Combine the original keys with the squared values to create a new dictionary
squares = dict(zip(nums.keys(), squared_values))

print(squares)
```

```python
#Lambda with sorted()

#Sorts data using a custom key.
data = [(1, 3), (4, 1), (2, 2)]
print(sorted(data, key=lambda x: x[1]))
```

Threading Vs Multiprocessing:

Threading → Running multiple threads inside the same program at the same time.

Multiprocessing → Running multiple processes (completely separate programs) at the same time.

```python
import threading

def task():
    print("Task running")

t1 = threading.Thread(target=task)
t2 = threading.Thread(target=task)

t1.start()
t2.start()

t1.join()
t2.join()
```
